# Исследование продаж компьютерных игр

Для интернет-магазина «Стримчик», продающего по всему миру компьютерные игры, необходимо провести анализ продаж компьютерных игр за определенный период времени (до 2016 года). Данные представлены из открытых источников в виде исторических данных о продажах игр, оценках их пользователей и экспертов, жанрах и платформах. На основе имеющейся информации необходимо выявить определяющие успешность игры закономерности, что позволит интернет-магазину выявлять потенциально наиболее успешный продукт и эффективно планировать проводимые рекламные кампании. Для этого сначала будет произведена преодобработак данных, после чего будет осуществлен исследовательский анализ данных, на основе чего составлен портрет потребителя компьютерных игр для каждого региона, также будет осуществлена постановка и проверка гипотез относительно игровых платформ и игровых жанров.

Данные в столбце user_score из формата object преобразовать в формат float, однако это невозможно из-за наличия в столбце значений tbd, что означает, что рейтинг игры не определен пользователями, возможно это связано с тем, что игра только вышла и указанный рейтинг по ней еще не сформирован, но в любом случае ввиду того, что данное значение отсутствует, заменяем его на nan

```python
df['user_score'] = df['user_score'].replace('tbd', np.nan)
```

## Проверка гипотез

Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные

Задайте самостоятельно пороговое значение alpha.
Поясните:
Как вы сформулировали нулевую и альтернативную гипотезы;
Какой критерий применили для проверки гипотез и почему.

Итак, осуществим проверку гипотезы о том, что средние пользовательские рейтинги платформ Xbox One и PC одинаковые


**H_0: средние пользовательские рейтинги платформ Xbox One и PC равны**


**H_a: средние пользовательские рейтинги платформ Xbox One и PC не равны**

```python
df_actual
```

```python
df_actual['user_score'].isna().sum()
```

```python
df_actual['platform'].unique()
```

```python
df_actual['user_score'].unique()
```

```python
user_score_xbox_one = df_actual[df_actual['platform'] =='xone']['user_score']
user_score_pc = df_actual[df_actual['platform'] =='pc']['user_score']
alpha = 0.05# alpha = значение уровня значимости

results = st.ttest_ind(user_score_xbox_one, user_score_pc, equal_var = False)# results = вызов метода для проверки гипотезы

print('p-значение:', results.pvalue.round(2))# вывод значения p-value на экран 

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") # условный оператор с выводом строки с ответом
```

***Вывод по итогу проверки гипотезы статистическим тестом: средние пользовательские рейтинги платформ Xbox One и PC равны***

Далее осуществим проверку гипотезы о том, что средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные


***H_0: средние пользовательские рейтинги жанров Action (англ. действие, экшен-игры) и Sports (англ. спортивные соревнования) равны***


***H_a: средние пользовательские рейтинги жанров Action (англ. действие, экшен-игры) и Sports (англ. спортивные соревнования) не равны***

```python
user_score_action = df_actual[df_actual['genre'] =='action']['user_score']
user_score_sports = df_actual[df_actual['genre'] =='sports']['user_score']
alpha = 0.05# alpha = значение уровня значимости

results = st.ttest_ind(user_score_action, user_score_sports, equal_var = False)# results = вызов метода для проверки гипотезы

print('p-значение:', results.pvalue)# вывод значения p-value на экран 

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") # условный оператор с выводом строки с ответом
```

***Вывод по итогу проверки гипотезы статистическим тестом: средние пользовательские рейтинги жанров Action (англ. действие, экшен-игры) и Sports (англ. спортивные соревнования) действительно разные***

Если говорить о том, каким образом были сформулированы нулевые и альтернативные гипотезы, проверенные в данном разделе анализа, то по общему правилу в качестве нулевой гипотезы мною было выбрано предположение именно о равенстве двух проверяемых значений выборок, соответственно в качестве альтернативной гипотезы было сформулировано положение об их неравенстве. В качестве порогового значения уровня значимости alpha, необходимого для сравнения с уровнем значимости p-value, мною был выбран также один из общепринятых пороговых значений в размере 5%

- резкое снижение количества новых игр после 2009 года;

- из 31 представленной в данных игровой платформы на 6 наиболее успешных приходится большая часть суммарной выручки за весь период, это такие платформы, как ps2, ps3, x360, ps, ds, wii, из распределения выручки по годам видно, что все платформы сначала показывают стабильный рост выручки, его пиковое значение, после чего выручка неизменно снижается, в среднем период существования платформ с наибольшим количеством выручки приходится на актуальный период с 1994 по 2016 год, именно к этому периоду, по-видимому, популярность компьютерных игр достигла существенных масштабов, о чем свидетельствует как раз резкое увеличение продаж игр, средний период "жизненного цикла" платформы составляет около 10 лет;

- с 2010 в данных наблюдается стабильное падение количества выпускаемых игр, поэтому последующие выводы будут формулироваться за период с 2012 года по 2016 год, то есть за последние 5 лет наблюдаемых данных;

- к лидирующим платформам по размеру суммарных продаж относятся ps3, ps4, x360, xone, 3ds, wiiu;

- платформы ps4 и xone можно сичтать потенциально прибыльными в дальнейшем, ввиду того, что на 2016 год их суммарные продажи были максимальными среди всех других платформ;

- средний размер суммарных продаж на всех платформах в актуальном периоде примерно на одном уровне, около 0.4 миллиона проданных копий;

- рассчитанная корреляция по нескольким платформам показала наличие взаимосвязи между показателями суммарных продаж и высокими оценками критиков, корреляция же между суммарными продажами и оценками пользователей здесь также не выявлена;

- наиболее популярным жанром игр в актуальном периоде является жанр action;

- жанры, приносящие наибольшие средние суммарные продажи: shooter, sports и platform (средние суммарные продажи более 0,20 млн копий), наименее продаваемыми жанрами являются жанры adventure и puzzle (средние суммарные продажи более 0,05 млн копий);

- в Северной Америке большую часть продаж (около 80 %, примерно по 20% каждая) берут на себя 4 платформы - ***x360, xone, ps3, ps4***, наиболее популярными являются жанр ***action*** (около 38 %);

- в Европе картина распределения популярности платформ и жанров практически аналогична американской, так наиболее популярными здесь являются платформы ***ps4*** (около 30 %) и ***ps3*** (около 25%), большую часть продаж (около 40 %) берет на себя также жанр ***action***;

- а вот в Японии несколько иной спиок лидирующих платформ: так лидером стала платформа ***3ds***, которая не была представлена в пердыдущих регионах,  далее с 27% идет ***ps3***, а в списке популярных жанров превалируют всего 2 жанра: снова ***action*** и ранее не встречавшийся ***role-playing*** (около 40 % и 30% соответственно);

- во всех трех регионах рейтинг ESRB не имеет существенного влияния на продажи.


Теперь перчислим характеристики, которые могут определять упешнность компьютерной игры:

- игры должны быть представлены на платформах ps4 и xone;

- по данным всех регионов потенциально более успешны игры с высокими оценками критиков, при этом оценки пользователей не имеют большого влияния на количество продаж;

- по данным всех регионов потенциально более успешны игры в жанрах shooter, sports и platform;

- при определении потенциально успешной игры в мировой практике такой показатель, как рейтинг ESRB, рассматривать не стоит.

- более подробная информация о потенциально успешных платформах и жанрах в разрезе регионов представлена выше.
